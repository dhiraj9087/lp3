```python
import pandas as pd
```

```python
df = pd.read_csv('sales_data_sample.csv', encoding = 'unicode_escape')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.drop(columns=['ORDERDATE','STATUS','QTR_ID','MONTH_ID','YEAR_ID'],inplace=True)
```

```python
df.head()
```

```python
df.drop(columns=['ADDRESSLINE2'],inplace=True)
```

```python
df['STATE'].fillna('Unknown', inplace=True)
df['POSTALCODE'].fillna('Unknown', inplace=True)
df['TERRITORY'].fillna('Unknown', inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
label_col=['PRODUCTLINE','PRODUCTCODE','COUNTRY','DEALSIZE']
from sklearn.preprocessing import LabelEncoder
for col in label_col:
  df[col] = LabelEncoder().fit_transform(df[col])
```

```python
label_col=['CUSTOMERNAME','ADDRESSLINE1','CITY','STATE','TERRITORY','CONTACTLASTNAME','CONTACTFIRSTNAME']
for col in label_col:
  df[col] = LabelEncoder().fit_transform(df[col])
```

```python
df['PHONE'] = LabelEncoder().fit_transform(df['PHONE'])
```

```python
df['POSTALCODE'] = LabelEncoder().fit_transform(df['POSTALCODE'])
```

```python
df.info()
```

```python
from sklearn.preprocessing import StandardScaler
data = StandardScaler().fit_transform(df)
```

```python
wcss=[]
from sklearn.cluster import KMeans
for k in range(1,15):
  kmeans = KMeans(n_clusters=k,init='k-means++',random_state=15)
  kmeans.fit(data)
  wcss.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, 15), wcss, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Inertia)')
plt.title('Elbow Method for Optimal k')
plt.grid()
plt.show()
```
